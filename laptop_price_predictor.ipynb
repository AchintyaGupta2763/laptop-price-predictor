```python
df10['CPU_Name'] = df10['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
df10.head()
```

```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Othert Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df10['CPU_Brand'] = df10['CPU_Name'].apply(fetch_processor)
```

```python
df10 = df10.drop(['Gpu','CPU_Name','Cpu'], axis = 'columns')
df10.head()
```

```python
def cat_os(inp):
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'
    elif inp == 'macOS'or inp == 'Mac OS X':
        return 'Mac'
    else:
        return 'Others/No OS/Linux'
```

```python
df10['OS'] = df10['OpSys'].apply(cat_os)
df10.head()
```

```python
df10 = df10[df10.Company != 'Xiaomi']
df10.head()
```

```python
df10 = df10.drop('OpSys', axis = 'columns')
df10.head()
```

```python
X = df10.drop('Price_in_rupees', axis = 'columns')
X.head()
```

```python
y=df10.Price_in_rupees
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
df10.head(1)
```

```python
from sklearn.linear_model import LinearRegression, Ridge,Lasso
```

```python
step1 = ColumnTransformer(transformers = [
    ('col_tnf', OneHotEncoder(sparse = False,drop='first'),[0,1,13,14])
],remainder='passthrough')
step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print('R2_score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
